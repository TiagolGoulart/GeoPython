```python
import geopandas as gpd
import matplotlib.pyplot as plt
import mplleaflet
```

```python
points_snis='/content/SNIS_2021_filtrado.gpkg'

points= gpd.read_file(points_snis)
```

```python
points['Código'] = points['Código'].astype(str)
```

```python
lixoes = points[points['UP003'] == 'Lixão']
aterro_controlado = points[points['UP003'] == 'Aterro controlado']
aterro_sanitario = points[points['UP003'] == 'Aterro sanitário']
```

```python
import folium

# Crie um mapa
m = folium.Map(location=[-15.788497,-47.879873], zoom_start=4)
```

```python
# Criando Layer Lixões

layer_lixoes = folium.FeatureGroup(name='Lixões')


# Itere sobre o GeoDataFrame
for index, row in lixoes.iterrows():
    # Crie um marcador de círculo
    folium.CircleMarker(
        location=[row['UP088 - lat.'], row['UP089 - lon.']],  # Use as colunas corretas para latitude e longitude
        radius=3,
        color='blue',
        fill_color='blue',
        fill_opacity=1,
        popup=folium.Popup(
            f"""<b>UP001:</b> {row['UP001']}<br>
                <b>Nome:</b> {row['Nome']}<br>
                <b>Código:</b> {row['Código']}"""

        ),
        zoom_on_click=True
    ).add_to(layer_lixoes)

layer_lixoes.add_to(m)
```

```python
# Criando Layer Aterro Sanitário

layer_aterro_sanitario = folium.FeatureGroup(name='Aterro Sanitário')


# Itere sobre o GeoDataFrame
for index, row in aterro_sanitario.iterrows():
    # Crie um marcador de círculo
    folium.CircleMarker(
        location=[row['UP088 - lat.'], row['UP089 - lon.']],  # Use as colunas corretas para latitude e longitude
        radius=3,
        color='yellow',
        fill_color='yellow',
        fill_opacity=1,
        popup=folium.Popup(
            f"""<b>UP001:</b> {row['UP001']}<br>
                <b>Nome:</b> {row['Nome']}<br>
                <b>Código:</b> {row['Código']}"""

        ),
        zoom_on_click=True
    ).add_to(layer_aterro_sanitario)

layer_aterro_sanitario.add_to(m)
```

```python
# Criando Layer Aterro Controlado

layer_aterro_controlado = folium.FeatureGroup(name='Aterro Controlado')


# Itere sobre o GeoDataFrame
for index, row in aterro_controlado.iterrows():
    # Crie um marcador de círculo
    folium.CircleMarker(
        location=[row['UP088 - lat.'], row['UP089 - lon.']],  # Use as colunas corretas para latitude e longitude
        radius=3,
        color='green',
        fill_color='green',
        fill_opacity=1,
        popup=folium.Popup(
            f"""<b>UP001:</b> {row['UP001']}<br>
                <b>Nome:</b> {row['Nome']}<br>
                <b>Código:</b> {row['Código']}"""

        ),
        zoom_on_click=True
    ).add_to(layer_aterro_controlado)

layer_aterro_controlado.add_to(m)
```

```python
# Adicionar o estilo de satélite do Google Hybrid como uma camada de tile
tile = folium.TileLayer(
    tiles='https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}',
    attr='Map data ©2015 Google, Dados do SNIS 2021',
    name='Google Satélite - Híbrido',
    max_native_zoom=24,
    overlay=True,
    control=True,
    show=False
).add_to(m)

folium.LayerControl(collapsed=False).add_to(m)
```

```python
from folium.plugins import MiniMap

# Crie um minimapa
minimap = MiniMap()
m.add_child(minimap)

# Exiba o mapa
m
```
