```python
# Import the libraries
import random
import matplotlib.pyplot as plt
```

```python
# First generate random data to demonstrate how to create boxplots
# Create lists to store the random numbers
a,b,c=[],[],[]
```

```python
# Generate the random numbers and save them in the lists created above
for i in range(1,100):
    a.append(random.randint(0,100))
    b.append(random.randint(50, 150))
    c.append(random.randint(-50, 50))
```

```python
# Create the figure and axis to plot the boxplots
fig, ax = plt.subplots()
# Plot the boxplots and set the labels
ax.boxplot([a,b,c], labels=['A','B','C'])
# Set the y-axis label
ax.set_ylabel('Distribuição de números aleatórios')
# View the chart
plt.show()
```
